```python
#1- Créer Un DataFrame en utilisant les données de fichier ‘’market_basket.txt’’ qui contient les identifiants des caddies et les produits associées.
import pandas as pd
url='https://drive.google.com/file/d/1gs4IESX9QncSq5GLW6m0ganfke3ZQsId/view?usp=sharing'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url,sep='\t')
#2-Afficher les 10 premières lignes du DataFrame.
df.head(10)
```

```python
#3-Afficher les dimensions du dataframe.
df.shape
```

```python
#4-Écrire un script python qui permet de Construire un table binaire indiquant la présence de chaque produit au niveau des caddies (True:1 si le produit est
# présent dans le caddie et 0 dans le cas réciproque).
f=df.pivot(values='Product' ,index='ID', columns='Product').fillna(value=0)
for i in range(len(f.axes[0])):
  for j in range(len(f.axes[1])):
    if f.iloc[i,j]!=0:
      f.iloc[i,j]=1
f
```

```python
#5-Tester la bibliothèque pandas.crosstab pour construire la table binaire et vérifier que vous avez les mêmes résultats de votre script.
d =pd.crosstab(df["ID"],df["Product"],)
```

```python
d
```

```python
#6-Afficher les 30 premières transactions et les 3 premiers produits.
d.iloc[:30,:3]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frq_items = apriori(d, min_support = 0.025, use_colnames = True)
  
frq_items
```

```python
rules = association_rules(frq_items, metric ="lift", min_threshold = 4)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head())
```

```python
#8-Afficher les 15 premiers itemsets.
frq_items.iloc[0:15]
```

```python
#9-Ecrire une fonction is_inclus() qui permet de vérifier si un sous-ensemble
#items est inclus dans l’ensemble x.
def is_inclus( x, items):
  return items.issubset(x)
print(is_inclus({'1','2','3'},{'1','2'}))
```

```python
#10-Afficher les itemsets comprenant le produit ‘Aspirin’.
import numpy as np
id = np.where(frq_items.itemsets.apply(is_inclus,items={'Aspirin'}))
print(frq_items.loc[id])
```

```python
#11-Afficher les itemsets contenant Aspirin et Eggs.
print(frq_items[frq_items['itemsets'].ge({'Aspirin','Eggs'})])
```

```python
#12-Nous produisons les règles à partir des itemsets fréquents.
regles = association_rules(frq_items,metric="confidence",min_threshold=0.75)
#13-Afficher les 5 premières règles.
print(regles.iloc[:5,:])
```

```python
#14-Filtrer les règles en affichant celles qui présentent un LIFT supérieur ou égal à 7.
liftRegles = regles.loc[:,['antecedents','consequents','lift']]
print(liftRegles[liftRegles['lift'].ge(7.0)])
```

```python
#15-Filtrer les règles en affichant celles menant au conséquent {‘2pct_milk’}.
print(liftRegles[liftRegles['consequents'].eq({'2pct_Milk'})])
```
